## Lesson 1 - Quiz 2, Excel to CSV

```python
import xlrd
import os
import csv
from zipfile import ZipFile

datafile = "2013_ERCOT_Hourly_Load_Data.xls"
outfile = "2013_Max_Loads.csv"


def open_zip(datafile):
    with ZipFile('{0}.zip'.format(datafile), 'r') as myzip:
        myzip.extractall()


def parse_file(datafile):
    workbook = xlrd.open_workbook(datafile)
    sheet = workbook.sheet_by_index(0)
    data = {}
    # process all rows that contain station data
    for n in range (1, 9): # there are 9 station columns.  Looping through each column field.    
        station = sheet.cell_value(0, n) #assigning the station name to each value in the top row (0). 'n' means no specified column range.
        cv = sheet.col_values(n, start_rowx=1, end_rowx=None) # stating that the sheet's values encompass all values startin from row 1.
        
        maxval = max(cv) # getting the max value as it iterates through each column 'n'.
        maxpos = cv.index(maxval) + 1 # gets the location of maxval + 1, since python starts at 0 but excel starts at 1. 
        maxtime = sheet.cell_value(maxpos, 0) # getting the 'Hour_end' time value; cell_value(row, column).
        realtime = xlrd.xldate_as_tuple(maxtime, 0) # takes an excel number expected to be a datetime and formats it follow Gregorian format.
        # for reference see:  http://xlrd.readthedocs.io/en/latest/api.html?highlight=cell_value#xlrd.sheet.Sheet.cell_value
        data[station] = {"maxval": maxval, 
                        "maxtime": realtime} # data is a dict, [station] is the key.  We are nesting a dict within a dict.
                                            # for each station key, we are adding the maxval and realtime values.  
    print data
    return data

def save_file(data, filename):
    with open(filename, "w") as f:
        w = csv.writer(f, delimiter='|')
        w.writerow(["Station", "Year", "Month", "Day", "Hour", "Max Load"])
        for s in data:
            year, month, day, hour, _ , _= data[s]["maxtime"]
            w.writerow([s, year, month, day, hour, data[s]["maxval"]])
    
```

```python
def get_author(root):
    authors = []
    for author in root.findall('./fm/bibl/aug/au'):
        data = {
                "fnm": None,
                "snm": None,
                "email": None
        }
        data["fnm"] = author.find('./fnm').text
        data["snm"] = author.find('./snm').text
        data["email"] = author.find('./email').text

        authors.append(data)

    return authors
```

```python
def get_authors(root):
    authors = []
    for author in root.findall('./fm/bibl/aug/au'):
        data = {
                "fnm": None,
                "snm": None,
                "email": None,
                "insr": []
        }
        data["fnm"] = author.find('./fnm').text
        data["snm"] = author.find('./snm').text
        data["email"] = author.find('./email').text
        insr = author.findall('./insr')
        for i in insr:
            data["insr"].append(i.attrib["iid"])
        authors.append(data)

    return authors
```

```python
def check_loc(point, lat, longi):
    # YOUR CODE HERE
    if point == 'NULL':
        point = None
    else:
        if point[0:1] == '{':
            point_len = len(point) - 1
            point_no_bracket = point[1:point_len]
            point_no_bracket.split('|')
            point_0 = point_no_bracket[0]
            point_1 = point_no_bracket[1]
        else:
            point.split()
            point_0 = point[0]
            point_1 = point[1]
            
    if point_0 == lat and point_1 == longi:
        return True
    else:
        return False
```
